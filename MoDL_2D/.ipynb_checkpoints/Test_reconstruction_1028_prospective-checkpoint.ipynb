```python
from skimage.metrics import structural_similarity as ssim
```

```python
def nmrse(x,gt):
    return np.linalg.norm(x-gt)/np.linalg.norm(gt)
def ssim_k(x,gt):
    return ssim(abs(x),abs(gt),data_range=abs(gt).max() - abs(gt).min())
def psnr(x,gt):
    mse = np.linalg.norm(x-gt)**2/(x.shape[0]*x.shape[1])
#     print(mse)
    return 20 * np.log10(abs(gt).max()/np.sqrt(mse)) 
def psnr_dynamic(x,gt):
    frames = gt.shape[-1]
    psnr_all = 0
    for j in range(frames):
        psnr_all += psnr(x[:,:,j],gt[:,:,j])
    return psnr_all/frames
def ssim_dynamic(x,gt):
    frames = gt.shape[-1]
    ssim_all = 0
    for j in range(frames):
        ssim_all += ssim_k(x[:,:,j],gt[:,:,j])
    return ssim_all/frames
```

```python
def compute_metrics(output,target_torch):
    cplx_error = cplx.abs(output - target_torch)
    cplx_l1 = torch.mean(cplx_error)
    cplx_l2 = torch.sqrt(torch.mean(cplx_error**2))
    cplx_psnr = 20 * torch.log10(scale / cplx_l2)
    return cplx_l1,cplx_l2,cplx_psnr
```
