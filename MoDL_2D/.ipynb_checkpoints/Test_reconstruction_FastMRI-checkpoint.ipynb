```python
%matplotlib notebook
import os, sys
import logging
import random
import shutil
import time
import argparse
import numpy as np
import sigpy.plot as pl
import torch
import torchvision
from tensorboardX import SummaryWriter
from torch.nn import functional as F
from torch.utils.data import DataLoader
from utils.flare_utils import torch2np
import h5py
from utils.resnet2p1d import generate_model
# import matplotlib
# matplotlib.use('TkAgg')
import sigpy as sp
# import custom libraries
from utils import transforms as T
from utils import subsample as ss
from utils import complex_utils as cplx
sys.path.append("models/unrolled3D/")

# import custom classes
from utils.datasets import SliceData
from utils.subsample import VDktMaskFunc
from models.unrolled3D.unrolled2D_MoDL import UnrolledModel
from subsample_fastmri import MaskFunc
# os.environ["CUDA_VISIBLE_DEVICES"]="9"
device = torch.device('cuda:3')
# device = torch.device('cuda:1' if torch.cuda.is_available() else 'cpu')
%load_ext autoreload
%autoreload 2
```

```python
da = h5py.File("/mikQNAP/NYU_knee_data/knee_train_h5/data/validate/329_7.h5", 'r')
kspace = np.array(da["kspace"])[()]
target = np.array(da["target"])[()]
mps = np.array(da["sensmaps"])[()]
kspace_torch = cplx.to_tensor(kspace).unsqueeze(0)
maps_torch   = cplx.to_tensor(mps).unsqueeze(0)
target_torch = cplx.to_tensor(target).unsqueeze(0)
mask_slice = np.ones((640,372))

# Create Sampling mask
train_mask = MaskFunc([0.08],[5])

mk1 = train_mask((1,1,372,2))[0,0,:,0]
knee_masks = mask_slice*mk1
mask_torch = torch.tensor(knee_masks[None,None,...,None]).float()  
kspace_torch = kspace_torch*mask_torch

input = kspace_torch.to(device)
maps = maps_torch.to(device)
```

```python
# Load the chzzeckpoint:
checkpoint_file = "/mikQNAP/NYU_knee_data/knee_train_h5/checkpoints/summary/train-3D_MELD_4steps_MoDLflag0_CGsteps_6date_20210330_ufloss0/model_epoch0.pt"
checkpoint = torch.load(checkpoint_file,map_location=device)
args = checkpoint['args']

if args.modl_flag == True:
    model = UnrolledModel(args).to(device)
model.load_state_dict(checkpoint['model'])
print(checkpoint['epoch'])
```

```python
with torch.no_grad():
    model.initiate(input,maps)
    start = time.time()
    output = model.forward()
    print(time.time()-start)
    
```
